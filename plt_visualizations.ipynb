Import libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

Import data set

```python
df = pd.read_csv('World_Bank_CO2_Data.csv')
```

```python
df.head()
```

```python
df.describe()
```

Remove NA values

```python
df = df.dropna()

df.head()
```

```python
df['Region'].unique()
```

```python
df_afg = df[df['Region'] == "South Asia"]
df_afg = df_afg[df_afg['Country Name'] == "Afghanistan"]
```

```python
df_afg.head()
```

```python
df_afg.plot(x="Year", y="CO2 Per Capita (metric tons)", kind='bar', figsize=(20,15))
```

```python
x = df['Region']
y = df['CO2 Per Capita (metric tons)']
plt.figure(figsize=(20,10))
plt.scatter(x, y, label = "CO2 Emissions")
plt.show()
```

```python
df_usa = df[df['Country Name'] == "United States"]
df_ru = df[df['Country Name'] == "Russian Federation"]
df_ch = df[df['Country Name'] == "China"]
df_afg = df[df['Country Name'] == "Afghanistan"]
```

```python
x = df_usa['Year']
y = df_usa['CO2 Per Capita (metric tons)']

x1 = df_ru['Year']
y1 = df_ru['CO2 Per Capita (metric tons)']

x2 = df_ch['Year']
y2 = df_ch['CO2 Per Capita (metric tons)']

x3 = df_afg['Year']
y3 = df_afg['CO2 Per Capita (metric tons)']

plt.figure(figsize=(20,10))
plt.plot(x, y, label = "USA")
plt.plot(x1, y1, label = "Russia")
plt.plot(x2, y2, label = "China")
plt.plot(x3, y3, label = "Afghanistan")

plt.legend()
plt.savefig("countries.png")
plt.show()
```

```python
df_sa = df[df['Region'] == "South Asia"]
```

```python
plt.figure(figsize=(20,10))

countries = df_sa['Country Name'].unique()

for country in countries:
    df_c = df[df['Country Name'] == country]
    x = df_c['Year']
    y = df_c['CO2 Per Capita (metric tons)']
    plt.plot(x, y, label = country)

plt.legend()
plt.savefig("SE_asia.png")
plt.show()
```
